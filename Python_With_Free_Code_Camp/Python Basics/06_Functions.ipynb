# Scope of variables

### LEGB scopes are lised below.
- L- Local Scope. \n
- E- Enclosing Scope (enclosed in a nested function).
- G- Global scope Variable defind at the top level.
- B- Built in Scope : Reserved names in python for predefined functions , modules, keywords, and objects.


```python
### Local Scope
def my_local_func():
    my_local_var = 10
    print(my_local_var)

my_local_func()
# print(my_local_var) , #we can't call out side the function
```

```python
### Enclosing Scope , nested inside the function athat access the variables of the function it's nested within
def outer_fun():
    msg = 'Hello there!'

    def inner_fun():
        print(msg)

    inner_fun()

outer_fun()
```

```python
### Example of Enclosed Scope
def outer_func1():
    msg = 'Hello there!'
    print(msg) # if we give the error message it will not function
    def inner_func():
        res = 'How are you?'
        print(msg)

    inner_func()
# inner_func()
outer_func1()
```

```python
def outer_func2():
    msg = 'Hello there!'
    res = "" ## we are making this non local scope
    def inner_func2():
        nonlocal res
        res = 'How are you ?'
        print(msg)

    inner_func2()
    print(res)

outer_func2()
```

```python
## Global Scope
my_var = 100
def show_var():
    my_var2 = '200'
    print(my_var)
    print(my_var2)
    return my_var2

show_var()
print(my_var)
# print(my_var2)
```

```python
my_var_1 = 7

def show_vars2():
    global my_var_2
    my_var_2 =10
    print(my_var_1)
    print(my_var_2)

show_vars2()
print(my_var_2)
```

```python
my_global_var1 = 10
def change_var():
    global my_var
    my_var = 20

change_var()
print(my_var)
```
