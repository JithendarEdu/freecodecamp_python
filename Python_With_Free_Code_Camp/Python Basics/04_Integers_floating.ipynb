# Working With Integers and floating

```python
# modulo operator
```

```python
my_int_1 = 56
my_int_2 = 12

my_float_1 = 5.4
my_float_2 = 12.0
mod_int = my_int_1 % my_int_2
mod_floats = my_float_1 % my_float_2

print('my modulus int is :', mod_int)
print('my modulus float is :', mod_floats)
```

```python
my_int_1 = 4.798
my_int_2 = 4.253

rounded_int_1 = round(my_int_1, 1)
rounded_int_2 = round(my_int_2 , 1)

print(rounded_int_1) # 5
print(rounded_int_2) # 4.3
```
